```python
import pandas as pd
import numpy as np
import ktrain
from ktrain import text
import tensorflow as tf
```

```python
!pip install openpyxl
```

```python
train_data = pd.read_excel("./IMDB-Movie-Reviews-Large-Dataset-50k/train.xlsx", dtype= str)
```

```python
test_data = pd.read_excel("./IMDB-Movie-Reviews-Large-Dataset-50k/test.xlsx", dtype=str)
```

```python
train_data.head()
```

```python
test_data.head()
```

Preprocessing on Traning and Test Dataset (use KTRAIN for that)

```python
(X_train, y_train),(X_test, y_test),preproc = text.texts_from_df(train_df=train_data,
                                                                 text_column = 'Reviews',
                                                                 label_columns = 'Sentiment',
                                                                 val_df = test_data,
                                                                 maxlen=500,
                                                                 preprocess_mode='bert')
```

```python
model = text.text_classifier(name='bert',
                             train_data = (X_train, y_train),
                             preproc = preproc,
                             )
```

```python
learner = ktrain.get_learner(model=model, train_data=(X_train, y_train),
                  val_data = (X_test, y_test),
                  batch_size=6)
```

**Finding out the best learning rate
--It may take days or many days to find out

```python
# learner.lr_find()
# learner.lr_plot()
```

```python
learner.fit_onecycle(lr = 2e-5, epochs = 1)
```

```python
predictor = ktrain.get_predictor(learner.model,preproc)
```

```python
data = ['Corresponds to the description. Tactile is pleasant. in quality it seems good, in operation I will check. The card does not read through it. 7 cards fit.(8 can be shoved, but the mechanism does not work)',
        'Interesting goods. Not very clearly works, but it works and even convenient if you get used to using. Delivery without questions. Packed well.',
        "The mini-thermos was without a temperature indicator. In addition, later, the seller told me that there are no green thermoses in the warehouse, you can find only black. In general, I'm happy. Everything else is in operation"]
```

```python
predictor.predict(data)
```
